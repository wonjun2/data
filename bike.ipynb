```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#데이터 불러오기

bike = pd.read_csv("C:/Users/cwj/Desktop/python/bike.csv")
 
bike.head()
```

```python
bike.info()
```

```python
bike['datetime'] = pd.to_datetime(bike['datetime'])
```

```python
bike['year'] = bike['datetime'].dt.year
bike['month'] = bike['datetime'].dt.month
bike['day'] = bike['datetime'].dt.day
bike['hour'] = bike['datetime'].dt.hour
 
bike.head()
```

```python
plt.style.use('seaborn')
sns.set(font_scale=2.5)
```

```python
fig, ax = plt.subplots(figsize=(16,12))

bike[['temp', 'atemp', 'humidity', 'windspeed']].hist(bins=50, ax=ax)
 
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16,14))

ax[0][0].scatter(bike['atemp'], bike['count'])
ax[0][1].scatter(bike['humidity'], bike['count'])
ax[1][0].scatter(bike['temp'], bike['count'])
ax[1][1].scatter(bike['windspeed'], bike['count'])

ax[0][0].set(title='atemp')
ax[0][1].set(title='humidity')
ax[1][0].set(title='temp')
ax[1][1].set(title='windspeed')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,7))

sns.boxplot(data=bike, x='holiday', y='count', ax=ax[0])
sns.boxplot(data=bike, x='workingday', y='count', ax=ax[1])
 
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,7))

sns.boxplot(data=bike, x='season', y='count', ax=ax[0])
sns.boxplot(data=bike, x='weather', y='count', ax=ax[1])

ax[0].set(xticklabels=['Spring', 'Summer', 'Fall', 'Winter'])
ax[1].set(xticklabels=['Sunny', 'Cloud', 'Light', 'Heavy'])
 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))

sns.barplot(data=bike, x='day', y='count', ax=ax, ci=None)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))

sns.barplot(data=bike, x='month', y='count', ax=ax, ci=None)

labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct','Nov', 'Dec']
ax.set(xticklabels=labels)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))

sns.pointplot(data=bike, x='hour', y='count',hue='season', ax=ax, ci=None)

labels=['Spring', 'Summer', 'Fall', 'Winter']
ax.legend(labels=labels)

plt.show()
```

```python
bike['day_name'] = bike['datetime'].dt.day_name()
 
fig, ax = plt.subplots(figsize=(20,8))

sns.pointplot(data=bike, x='hour', y='count',hue='day_name', ax=ax, ci=None)

plt.show()
```
